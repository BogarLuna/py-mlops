```python
from typing import Dict
import requests
import pandas as pd
def get_data_df(url:str,params:Dict={}):
    response= requests.get(url).json()
    print(response.get('info'))
    num_pages = response.get('info').get('pages')
    data = pd.DataFrame()
    if num_pages:
        for page in range(1, num_pages+1):
            response= requests.get(url,params={'page': page}).json()
            delta = pd.DataFrame(data= response.get('results'))
            data = pd.concat([data,delta],ignore_index=True)
    return data
```

```python
rick=get_data_df('https://rickandmortyapi.com/api/episode')
```
